```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.figure(figsize=(4,3), dpi=100)

plt.plot(x, y, 'b^--', label='2*x')

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')

plt.title('Our First Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 10})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])

plt.legend()
plt.show()
```

```python
labels=['A','B','C']
values=[1,4,2]

bars = plt.bar(labels,values)

bars[0].set_hatch('/')

plt.figure(figsize=(6,4))

plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)')

plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollar')

plt.legend()
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head()
```

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]

right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f')
plt.title('Foot Preferece of FIFA Player', fontdict={'fontsize':10})

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

fifa.Weight[0]
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4, .1, 0, 0, .4)

plt.style.use('ggplot')
plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.show()
```

```python
barcelone = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']

madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']

revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']

plt.figure(figsize=(6,8))
plt.style.use('default')
boxes = plt.boxplot([barcelone, madrid, revs], labels=labels, patch_artist=True, medianprops={'linewidth':4})

for box in boxes['boxes']:
    box.set(color='#4286f4', linewidth=2)
    
    box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison', fontdict={'fontsize':10})
plt.ylabel('FIFA Overall Rating')

plt.show()
```
